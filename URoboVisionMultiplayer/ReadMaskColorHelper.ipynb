## Find Mask Color Bson File in given folder
* Input bson file path and return file names

```python
## give the correct bson file folder path here
file_path = "E:/Work_AI/test/RobCoG/Vision Logger/2018_8_10_6_44/Cam01/Mask Color/"
```

```python
def read_bson_file(file_path):
    file_names = []
    if os.path.isdir(file_path):
        for file in os.listdir(file_path):
            print(file)
            if file.endswith(".bson"):
                file_name=file
                file_names.append(file_name)
        
        if len(file_name):
            print("Read data from bson file")
            
        else:
            print("Don't contain any bson file in this folder")
            
    else:
        print("The given folder doesn't exist.Please give correct folder path")
    return file_names
```

## Save Mask color as dictionaries

```python
def save_images_from_bson(FilePath,FileNames):
    Mask_Color_Dict={}
    if len(FileNames):
        for file in FileNames:
            data = bson.decode_file_iter(open(FilePath+file,'rb'))
            while True:
                try:
                    data_dict = next(data)
                    for mask_data in data_dict.values():
                        Mask_Color_Dict[file[:-5]]=mask_data
                except:
                    print("Mask color in {} has been saved in dictionary".format(file))
                    break
    return Mask_Color_Dict
```

## Final Function to do task  above

```python
def Read_Save_Image_in_Bson(file_path):
    filenames = read_bson_file(file_path)
    Mask_dict=save_images_from_bson(file_path,filenames)
    return Mask_dict
```
